```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Função para gerar datas aleatórias
def random_dates(start, end, n):
    start_u = start.value // 10**9
    end_u = end.value // 10**9
    return pd.to_datetime(np.random.randint(start_u, end_u, n), unit='s')

# Definindo o intervalo de datas
start_date = pd.to_datetime('2022-01-01')
end_date = pd.to_datetime('2022-12-31')

# Gerando dados fictícios
data = {
    'Data de venda': random_dates(start_date, end_date, 40),
    'Nome vendedor': [random.choice(['Alice', 'Bob', 'Carlos', 'Diana', 'Eva']) for _ in range(40)],
    'Quantidade vendida': np.random.randint(1, 20, 40),
    'Preço de venda': np.round(np.random.uniform(10, 100, 40), 2),
    'Custo unitário': np.round(np.random.uniform(5, 50, 40), 2),
    'Produto': [random.choice(['Produto A', 'Produto B', 'Produto C', 'Produto D']) for _ in range(40)],
    'Cidade': [random.choice(['Rio de Janeiro', 'São Paulo', 'Belo Horizonte', 'Porto Alegre', 'Curitiba']) for _ in range(40)],
    'Despesa de frete com a venda': np.round(np.random.uniform(5, 20, 40), 2)
}

# Criando o DataFrame
df = pd.DataFrame(data)

# Calculando a receita de vendas e o custo do produto vendido
df['Receita de vendas'] = df['Quantidade vendida'] * df['Preço de venda']
df['Custo do produto vendido'] = df['Quantidade vendida'] * df['Custo unitário']

# Exibindo o DataFrame
print(df)

# Salvando o DataFrame em um arquivo CSV
df.to_csv('base_vendas.csv', sep=',', index=False)
```
